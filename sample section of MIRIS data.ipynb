```python
import numpy as np
from astropy import visualization
import matplotlib.pyplot as plt
from background_per_pix import miris_bgd_flux
```

# 7/11/2021
This is an attempt to get a small 10 by 10 region of MIRIS data for comparison to the original. I picked this location because it looked vaguely interesting.

```python
scale = 47.5/3600
glon_and_glat = [(11.35 + scale*ii, 0.62 + scale*jj) for ii in range(10) for jj in range(10)]
```

```python
table_f = miris_bgd_flux(glon_and_glat)
```

```python
fig, ax = plt.subplots(figsize=(5,4))
sc = ax.scatter(table_f['longitude'], table_f['latitude'], c=table_f['flux paa'], marker="s", 
                cmap='afmhot', s=510)
ax.set_ylabel('Galactic Latitude [Deg]', fontsize='10')
ax.set_xlabel('Galactic Longitude [Deg]', fontsize='10')
plt.ylim(0.6167, 0.743)
plt.xlim(11.345, 11.4745)
plt.title('Background Flux in a Small Region', fontsize='15')
cbar = fig.colorbar(sc)
cbar.set_label("Flux per pixel [Jy]", fontsize='10')
```
