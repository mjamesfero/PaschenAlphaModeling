```python
from astropy import visualization
import matplotlib.pyplot as plt
from MIRIS_flux_calculator import ks_mags, fluxes
```

```python
paa = [flux.value for flux in fluxes['paa']]
paach = [flux.value for flux in fluxes['paach']]
paacl = [flux.value for flux in fluxes['paacl']]
```

```python
fig, ax = plt.subplots(figsize=(15,5))
sc = ax.scatter(ks_mags, paa, s=200, c='b')
sc = ax.scatter(ks_mags, paach, s=200, c='r')
sc = ax.scatter(ks_mags, paacl, s=200, c='g')
ax.set_ylabel('Flux (Jy)', fontsize='20')
ax.set_xlabel('Ks Magnitude', fontsize='20')
plt.title('Color Correction for Different Magnitudes', fontsize='30')
```
